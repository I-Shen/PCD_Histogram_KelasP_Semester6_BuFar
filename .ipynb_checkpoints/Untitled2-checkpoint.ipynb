```python
#IMPORT LIBRARY
import cv2
import numpy
import glob
import matplotlib.pyplot as plt
from scipy import ndimage, misc

#BUKA IMAGE DALAM FOLDER MASUKKAN KEDALAM LIST
folders = glob.glob('C:/Users/NUGI/UAS_PCD/archive/plate_number_recognition_datasets_v1/images/train')
imagenames_list = []
for folder in folders:
    for f in glob.glob(folder+'/*.jpeg'):
        imagenames_list.append(f)
```

```python
#FUNGSI MENJADIKAN IMAGE RGB KEDALAM GRAYSCALE
gray = []

for img in imagenames_list:
    RGB = cv2.imread(img)
    img2 = 0.02126*RGB[:,:,2] + 0.7152*RGB[:, :,1] + 0.0722*RGB[:,:,0]
    gray.append(img2)

plt.imshow(gray[3], cmap='gray')
```

```python
#TAMBAHAN DARI SINI
#FUNGSI MENJADIKAN IMAGE GRAYSCALE KEDALAM BENTUK THRESHOLD ATAU PROSES BINARIZATION
binarization = []

for binar in gray:
    retval, threshold = cv2.threshold(binar, 170, 255, cv2.THRESH_BINARY_INV)
    binarization.append(threshold)

plt.imshow(binarization[3], cmap='gray')
```

```python
#FUNGSI MENJADIKAN IMAGE THRESHOLD MENJADI BLUR KARENA PROSES MEDIAN FILTER
mefil = []

for median in binarization:
#     blur = cv2.medianBlur(median, 5)
    blur = ndimage.median_filter(median, size=2)
    mefil.append(blur)

plt.imshow(mefil[3], cmap='gray')
```
